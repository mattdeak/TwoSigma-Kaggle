```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
with pd.HDFStore('train.h5','r') as train:
    df = train.get("train")
```

```python
corr = df.corr()
corr_list = []
for i in range(len(corr.columns)):
    for j in range(i,len(corr.columns)):
        if np.abs(corr.ix[i,j]) > 0.8 and i != j:
            corr_list.append([i,j])
```

```python
fig = plt.figure()
dim_rows = int((np.sqrt(len(corr_list)+1)) + 1)
dim_cols = int(np.sqrt(len(corr_list)))
counter = 0
for i in range(len(corr_list)):
    coors = corr_list[i]
    ax = plt.subplot2grid((2,2),(0,i%2))
    x = df.ix[:,corr_list[i]]
    ax.scatter(x.ix[:,0],x.ix[:,1])
    plt.xlabel(x.columns[0])
    plt.ylabel(x.columns[1])
    plt.tick_params(labelbottom='off',labelleft='off')
    ax.set_title('Correlation = {:.3f}'.format(corr.ix[coors[0],coors[1]]))
    if (i%2 == 0 and i != 0) or i == len(corr_list)-1 :
        plt.show()


    
    
```
